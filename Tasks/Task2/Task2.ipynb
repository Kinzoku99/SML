```python
import pandas as pd
import numpy as np
import plotly.express as px
```

# Exploration

```python
X_train = pd.read_csv('X_train.csv', sep=';')
X_test = pd.read_csv('X_test.csv', sep=';')
y_train = pd.read_csv('y_train.csv', sep=';')
```

```python
print(f"There are {X_train.shape[0]} observations in the training data.")
assert y_train.shape[0] == X_train.shape[0] and X_train.shape[0] == 3794
print(f"There are {X_test.shape[0]} observations in the test data.")
assert X_test.shape[0] == 660
print(f"There are {X_train.shape[1]} variables in the training data.")
assert X_test.shape[1] == 9000  and y_train.shape[1] == 1
print(f"There are {X_test.shape[1]} variables in the test data.")
number_of_nulls = np.sum(X_train.isnull().sum().values)
number_of_nulls += np.sum(y_train.isnull().sum().values)
number_of_nulls += np.sum(X_test.isnull().sum().values)
assert number_of_nulls == 0, "Data is not complete!"
print("Data is complete for sure.")
```
